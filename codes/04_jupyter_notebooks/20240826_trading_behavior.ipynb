```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/01_original_data/swap_data/all_swaps.csv", low_memory=False)
```

```python
user_df = pd.read_stata("~/Downloads/without_fee.dta")
```

```python
lps = user_df["liquidity_provider"].unique()
df_filterd = df[df["recipient"].isin(lps)].copy()
```

```python
df_filterd["block_timestamp"] = pd.to_datetime(df_filterd["block_timestamp"])
df_filterd["week"] = df_filterd["block_timestamp"].dt.to_period('W-SAT').dt.start_time
res_temp = df_filterd.groupby(["recipient", "week", "pool_address"])["tx_hash"].nunique()
```

```python
res_temp = res_temp.reset_index()
```

```python
res_temp["liquidity_provider"] = res_temp["recipient"]
```

```python
temp_res = user_df.merge(res_temp, on=["liquidity_provider", "week", "pool_address"], how="left")
```
